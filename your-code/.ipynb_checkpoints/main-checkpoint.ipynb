## Reading the data

In the next cell, use Python to obtain data from the API endpoint you chose in the previous step. Workflow:

1. Make request to obtain data of the 1st page. Flatten the data and store it in an empty object variable.

```python
# your code here
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
from urllib.request import urlopen
```

```python
url='https://www.zalando.fr/api/catalog/articles?categories=promo-femme&limit=84&offset=84'
```

```python
response = urlopen(url).read().decode('utf-8')
response

#formato json
results = json.loads(response)
results

#aplanas json en panda
flattened_data = pd.json_normalize(results)

#aplanas el json con los articulos en un panda
flattened_data1 = pd.json_normalize(flattened_data.articles[0])
flattened_data1
```

```python
#"El numero de paginas está dentro del json results
#dentro de la llave pagination, esta la llave page_count, que es lo que traemos
results
```

```python
total_pages=results['pagination']['page_count']
total_pages
```

```python
#Creamos df vacio
df_fem = pd.DataFrame()

#Creamos loop para traer articulos d todas las pags
#tarda mucho en correr, por lo que dejaremos 10 paginas
for i in range (int(total_pages/10)):
    
    #cada pagina tiene 84 articulos, la llave per_page nos dice eso
    k = 84*i
    #el url con la k es para ir cambiando de pagina
    url = f'https://www.zalando.fr/api/catalog/articles?categories=promo-femme&limit=84&offset={k}&sort=sale'
    #urlopen, el request no jala
    response = urlopen(url).read().decode('utf-8')
    #pasar a json
    results = json.loads(response)
    #aplanar json en un pd df
    flattened_data = pd.json_normalize(results)
    #aplanar el json que tiene los articulos en un pd df
    flattened_data1 = pd.json_normalize(flattened_data.articles[0])
    #sku como indice
    flattened_data1 = flattened_data1.set_index('sku')
    #agregamos el df que se genera en cada pagina a un df maestro, el primero que hicimos vacío
    df_fem = df_fem.append(flattened_data1)


df_fem
```

## Bonus

Extract the following information from the data:

* The trending brand.

* The product(s) with the highest discount.

* The sum of discounts of all goods (sum_discounted_prices divided by sum_original_prices).

```python
# your code here
#trending brand
#cuenta el numero de veces que aparece brand_name y trae el que más veces aparezca, el primero de la lista
df_fem.brand_name.value_counts().index[0]
```

```python
#The product(s) with the highest discount
#quitamos lo que no es digito
df_fem['price.original']=df_fem['price.original'].str.extract('(\d*,\d*)')
df_fem['price.promotional']=df_fem['price.promotional'].str.extract('(\d*,\d*)')

#cambiamos , por . para precios mexicanos
df_fem['price.original'] = [x.replace(',', '.') for x in df_fem['price.original']]
df_fem['price.promotional'] = [x.replace(',', '.') for x in df_fem['price.promotional']]
```

```python
df_fem['discount_amount']=df_fem['price.original'].astype(float)-df_fem['price.promotional'].astype(float)
df_fem1=df_fem.copy()
df_fem1
```

```python
total_disc=df_fem1.groupby(['name']).sum().discount_amount
```

```python
total_disc.sort_values(ascending=False)
```
